#### 1. 데이터에서 제주도 지역 데이터 추출하기
- 제주도 '행정기관' 내용을 이용해 '읍면동' 필드 생성   
  읍면동이 없는 데이터는 상위 구분(제주특별자치도/서귀포시)으로 입력
- 제주시와 서귀포시의 인구 현황 분석(어디가 더 많이 거주하는지 확인)-그래프로도 표현
- 제주시의 지역(읍면동 단위) 인구현황을 계산하고 막대 차트로 출력하시오

```python
# 파일 읽어오기
import pandas as pd

# 나이별 데이터
age_df = pd.read_csv("202202_age.csv", encoding="cp949", header = 3, thousands = ",")
display(age_df)    # 데이터 확인
print(age_df.info())      # 전체 정보확인
```

```python
# 행정기관에서 "제주시"이 포함되어 있는 데이터 전체를 조회후 df1에 저장
df1 = age_df[age_df['행정기관'].str.contains("제주특별자치도")]  # if "제주시" in row[1]:
df1
df1.iloc[[0]]
df1.info()
```

```python
display(df1)
```

```python
# 제주도 '행정기관' 내용을 이용해 '읍면동' 필드 생성
# 읍면동이 없는 데이터는 상위 구분(제주특별자치도/서귀포시)으로 입력

for row in df1.index:
    if df1.loc[row, '행정기관'].split(" ")[-2] == "":
        df1.loc[row,'읍면동'] = df1.loc[row, '행정기관'].split(" ")[-3]
    elif df1.loc[row, '행정기관'].split(" ")[-1] == "":
        df1.loc[row,'읍면동'] = df1.loc[row, '행정기관'].split(" ")[-2]
    else:
        df1.loc[row,'읍면동'] = df1.loc[row, '행정기관'].split(" ")[-1]
display(df1)
```

```python
# 전체인구현황 = tot_df
# 남자인구현황 = male_df
# 여자인구현황 = female_df
tot_df = []
male_df = []
female_df = []

tot_df = df1.iloc[:, 4:105:] 
male_df = df1.iloc[:, 107:208:] 
female_df = df1.iloc[:, 210:311:] 

```

```python
# 0세부터 100세이상 데이터만 출력(전체, 남자, 여자)

display(tot_df)
display(male_df)
display(female_df)
```
