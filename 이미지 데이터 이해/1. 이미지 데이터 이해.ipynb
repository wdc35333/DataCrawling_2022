```python
# pandas 모듈을 이용한 배열 이해
import pandas as pd
```

```python
cv2.split(img1)
```

```python
# 픽셀 배열을 B, G, R 단위 분할
b, g, r = cv2.split(img1)

# B의 값을 이용해 데이터 프레임 형식으로 변경
b_df = pd.DataFrame(b)
b_df

# 데이터 사이즈 확인
print(b_df.shape)

# 데이터 확인
display(b_df.head())
```

```python
# G의 값을 이용해 데이터 프레임 형식으로 변경
g_df = pd.DataFrame(g)

print(g_df.shape)

display(g_df.head())
```

```python
# R의 값을 이용해 데이터 프레임 형식으로 변경
r_df = pd.DataFrame(r)

print(r_df.shape)

display(r_df.head())
```
