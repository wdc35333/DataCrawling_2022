## * 정규화란?
- 머신러닝 알고리즘은 데이터가 가진 feature(특성)들을 비교하여 데이터의 패턴을 찾음
- 머신러닝에서 데이터가 가진 feature의 스케일이 심하게 차이가 나는 경우 문제가 될수 있음
- 정규화의 목적: 데이터셋의 numerical value 범위의 차이를 왜곡하지 않고 공통 척도로 변경하는 것
- 피처의 범위가 다른 경우에만 필요
- 데이터를 정규화 하는 두 가지 방법
 * Min-Max Normalization (최소-최대 정규화)
 * Z-Score Normalization (Z-점수 정규화)
- 선형회귀, 로지스틱 회귀, 서포트 벡터 머신 등의 알고리즘에서는 효과가 좋지만, 결정 트리 알고리즘에서는 효과가 없다.

#### 1. Min-Max Normalization (최소-최대 정규화)
- 최소-최대 정규화는 데이터를 정규화하는 가장 일반적인 방법이다. 모든 feature에 대해 각각의 최소값 0, 최대값 1로, 그리고 다른 값들은 0과 1 사이의 값으로 변환하는 것
- 계산방법(수학공식)
 * (X - MIN) / (MAX-MIN) 
 * 수학공식 표현

 image.png

- 최소-최대 정규화의 치명적인 단점: 이상치(outlier)에 너무 많은 영향을 받는다.
-
- (예)100개의 값이 있는데 그 중 99개는 0과 40 사이에 있고, 나머지 하나가 100이면 어떨까. 그러면 99개의 값이 모두 0부터 0.4 사이의 값으로 변환된다.
- 차트로 확인한 결과
 image.png


```python
def min_max_normalize(lst):
    normalized = []
    
    for value in lst:
        normalized_num = (value - min(lst)) / (max(lst) - min(lst))
        normalized.append(normalized_num)

    return normalized
```

#### 2. Z-Score Normalization (Z-점수 정규화)
- 표준화(Standardization)
- 표준정규분포의 속성을 갖도록 피처가 재조정되는 것
- 이상치(outlier) 문제를 피하는 데이터 정규화 전략
- 계산방법(공식)
 * (X - 평균) / 표준편차
 * 수학공식 표현

- 차트로 확인한 결과

- 여전히 찌그러져 보이긴 하지만, 그래도 거의 모든 점들이 x축과 y축에서 -2와 2 사이에 있어 어느정도 비슷한 스케일로 나타낸 것으로 보임

```python
def z_score_normalize(lst):
    normalized = []
    
    for value in lst:
        normalized_num = (value - np.mean(lst)) / np.std(lst)
        normalized.append(normalized_num)
    
    return normalized
```

### Normalization vs Standardization 정리

- 최소 최대 정규화: 모든 feature들의 스케일이 동일하지만, 이상치(outlier)를 잘 처리하지 못한다.
- Z-점수 정규화 : 이상치(outlier)를 잘 처리하지만, 정확히 동일한 척도로 정규화 된 데이터를 생성하지는 않는다.
