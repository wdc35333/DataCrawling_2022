- 조건을 이용한 데이터 추출
      단일조건: df[df["필드명"] > 기준값] / df[df.필드명 > 기준값]
      다중조건: df[(조건식1) & (조건식2)] / df[(조건식1) | (조건식2)] (&: AND연산 / |: OR연산)

```python
df1[df1.C >= 0]    # c열의 값이 0보다 큰 값
```

```python
# 다중 조건에 대한 AND(&) 연산
df1[(df1.B >= 0) & (df1["C"] >= 0)]    # b열과 c열의 값이 모두 0보다 큰 값
```

```python
# 다중 조건에 대한 OR(|) 연산
df1[(df1.B < 0) | (df1["C"] > 0)]  # B열이 0보다 작은값 또는 C열이 0보다 큰값 모두 출력
```

- 데이터에서 NaN 값 제거
 * df.dropna(): NaN이 1개라도 있는 행 모두 제거
 * df.dropna(axis=1): NaN이 1개라도 있는 열 삭제
 * df.dropna(how="all"): 행 전체가 NaN이면 해당행 삭제
 * df.dropna(thresh=N개): 각 행에 대하여 N개 이상의 값을 가지는 데이터(행)만 출력 
 * df.dropna(subset=["열이름", "열이름",..]): 지정한 열의 값이 NaN이면 해당 행 삭제

```python
df3.dropna()
```

```python
df3.dropna(axis=1)
```

[미션] new_val 값 생성
1. new_val 변수 생성
2. new_val 변수에 평균휘발유값-각 행별 휘발유값 추가
3. df3["평균대비"] 필드 생성 후 new_val 값을 대입

```python
# "평균대비" 필드에 추가
# df["새로운필드명"] = [값], [값]은 df의 데이터 갯수와 일치

# new_val 값 생성(각자 만들어보세요)
# 한줄쓰기로도 변경해보세요

new_val=[]
for i in df3.index:
    new_val.append(df3.loc[i, "휘발유"]-mean1)

df3["평균대비"] = new_val
df3.head()
```

```python
# new_val 값 생성(각자 만들어보세요)
del df3["평균대비"]

new_val=[df3.loc[i, "휘발유"]-mean1 for i in df3.index]

df3["평균대비"] = new_val
df3.head()
```

- pandas 데이터 병합
 * DataFrame 병합: merge()

```python
df1 = pd.DataFrame({'고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
                    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']}, 
                    columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({'고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
                    '금액': [10000, 20000, 15000, 5000, 100000, 30000]}, 
                    columns=['고객번호', '금액'])
df2
```

```python
# df1, df2의 공통열을 기준으로 데이터를 찾아서 공통으로 존재한 데이터(고객번호)만 연결
pd.merge(df1, df2)  
```

```python
# df1, df2의 모든 데이터를 출력
pd.merge(df1, df2, how="outer")  
```

```python
# df1, df2에서 왼쪽에 위치한 데이터를 기준으로 출력
pd.merge(df1, df2, how="left") 
```

```python
# df1, df2에서 오른쪽에 위치한 데이터를 기준으로 출력
pd.merge(df1, df2, how="right") 
```

- DataFrame 데이터 연결(데이터를 기존 데이터의 아래에 추가할 때)
 * df.concat([df1, df2, df3, ...]) : 열 제목을 기준으로 아래에 추가
 * df.concat([df1, df2, df3, ...], axis=1) : index 값을 기준으로 뒤(오른쪽)에 붙여준다.

```python
df1=pd.DataFrame({"이름":["홍길동", "이순신", "김파이", "이기준"], 
                  "나이":[30, 45, 29, 48], 
                  "성적":[99, 88, 77, 66]})
                 
df1
```

```python
df2=pd.DataFrame({"이름":["박노트", "이상식", "조규식", "서준석"], 
                  "나이":[74,88,68,59], 
                  "성적":[88,77,88,92]})
                 
df2
```

```python
df3=pd.DataFrame({"이름":["박지정", "김규식", "홍규식", "차준석"], 
                  "나이":[84,78,98,79], 
                  "성적":[98,87,78,82],
                  "비고":["간", "페", "위", "간"]})                
df3
```

```python
## 데이터의 열 이름이 동일한 경우 
pd.concat([df1, df2])
```

```python
# 필드명 다른 데이터 추가
pd.concat([df1, df3])
```

```python
# concat함수는 두개 이상의 DataFrame을 한번에 추가할 수 있다.
pd.concat([df1, df2, df3])
```

```python
### concat 열 추가
df1=pd.DataFrame([["홍길동", 25, "남"], ["이순신", 48, "남"], ["유관순", 18, "여"]],
                 columns=["이름", "나이", "성별"])
df1
```
