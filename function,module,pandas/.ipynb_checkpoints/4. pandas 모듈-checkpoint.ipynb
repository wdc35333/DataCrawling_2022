```python
##  행/열 데이터 출력
##  열단위 데이터
df1.columns   # 열 이름 확인하기
df1['휘발유']
df1[["경유","휘발유"]]    # 두 개 이상의 열값 출력 => []로 묶어서 입력
display(df1[["상표", "셀프여부", "경유", "휘발유"]])
df2=df1[["상표", "셀프여부", "경유", "휘발유"]]
```

```python
## 기술통계
df2.describe()
```

```python
## 행 기준 데이터 출력
# loc[], iloc[]
df1.loc[0]    # index 열 기준
df1.loc[1:3, ["휘발유","경유"]]
df2 = df1.loc[1:10, ["셀프여부","휘발유","경유", "상표"]]
df2
```

```python
## iloc[] : 데이터의 행/열 인덱싱된 값을 기준으로 출력
df2.iloc[0]
df2.iloc[1:5, 0]
df2.iloc[1:5, 1:]
df2.iloc[1:5, [0, 1, 3]]
```

- df2 경우/휘발유 값을 정수형 데이터로 변경
      df.dtypes: 각 열별 데이터 형식을 출력

```python
# df2.dtypes
# df2.index

lst = []
for i in df2.index:
    lst.append(int(df2.loc[i, ["휘발유"]]))   # 휘발유 열 값을 정수형 데이터로 변경 후 lst에 추가

print(lst)

df2["휘발유"] = lst   # 변경된 lst를 이용해 df2의 휘발유 열 값 업데이트

df2.dtypes
```

###[미션] 위 코드의 for 구문을 한줄쓰기로 변경(경유)

```python
lst2 = [int(df2.loc[i, ["경유"]]) for i in df2.index]
print(lst2)
```

### df3 생성후 경유(정수형)/휘발유(실수형) 데이터
- df.astype(데이터형, inplace = True), 모든 데이터를 입력한 데이터형으로 변경
- df = df.astype({"필드명": int, "필드명": float})

```python
df3 = df1.loc[1:10, ["셀프여부","휘발유","경유", "상표"]]
print(df3.dtypes)
print(len(df3))
```

```python
df3 = df3.astype({"경유":int, "휘발유":float})
print(df3.dtypes)
```

### [미션] df1의 데이터 형식 변경
- for문을 이용해 "휘발유", "경유" 값을 int형으로 변경
- 변경 값이 "-"이면 해당 값은 0 값으로 변경
- df1에 적용 저장

```python
# for문/if문을 이용한 데이터 변경 및 형식 변경
```

```python
## replace()
df1["휘발유"].replace("-", 0, inplace=True)
print(df1.휘발유)
df1=df1.astype({"휘발유":int})
print(df1.dtypes)
```
